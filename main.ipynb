```python
# Splitting the dataset
X = data.drop('Outcome', axis=1azaz)
Y = data['Outcome']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=10)
```

```python
# check columns with zero values

print("Total Rows : {}", format(len(data)))
print("No. of rows missing values Pregnancies : {}", format(len(data.loc[data['Pregnancies']==0])))
print("No. of rows missing values Glucose : {}", format(len(data.loc[data['Glucose']==0])))
print("No. of rows missing values BloodPressure : {}", format(len(data.loc[data['BloodPressure']==0])))
print("No. of rows missing values SkinThickness : {}", format(len(data.loc[data['SkinThickness']==0])))
print("No. of rows missing values Insulin : {}", format(len(data.loc[data['Insulin']==0])))
print("No. of rows missing values BMI : {}", format(len(data.loc[data['BMI']==0])))
print("No. of rows missing values DiabatesPedigreeFunction : {}", format(len(data.loc[data['DiabetesPedigreeFunction']==0])))
print("No. of rows missing values Age : {}", format(len(data.loc[data['Age']==0])))
```

```python
# imputing zero values in the dataset
from sklearn.impute import SimpleImputer

fill_values = SimpleImputer(missing_values=0, strategy='mean')
X_train = fill_values.fit_transform(X_train)
X_test = fill_values.fit_transform(X_test)
```

```python
# building model using Random Forest
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(X_train, Y_train)
```

```python
predictions = rfc.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy Score : ", format(metrics.accuracy_score(Y_test, predictions)))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(Y_test, predictions))
print(classification_report(Y_test, predictions))
```

```python
print("Prediction Probabilities : ")
rfc.predict_proba(X_test)
```
